```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
url = "https://conversationstartersworld.com/250-conversation-starters/"
```

```python
response = requests.get(url)
```

```python
soup = bs(response.text, "html.parser")
```

```python
ptags = soup.find_all("p")
```

```python
questions = [x.text for x in ptags]
```

```python
df = pd.DataFrame({
"Question": questions
})
df.head(10)
```

```python
df.drop(df.index[[0,1,2,3,4,5,284,285,286,287,288,289,290]], inplace=True)
df.head()
```

```python
df[["col1","col2","col3","col4"]] = df["Question"].str.split(".", expand=True)
```

```python
df.set_index("col1", inplace=True)
df.head()
```

```python
del df["Question"]
df.head()
```

```python
del df["col3"]
del df["col4"]
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```
